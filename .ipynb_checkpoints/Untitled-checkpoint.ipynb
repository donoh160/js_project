```python
import json
from countryinfo import CountryInfo
```

```python
# Sample Country 
country = CountryInfo('Andorra')
info = country.info()
info['name']
```

```python
# Load country json
# Includes only countries recognized by the US Department of State
f = open('countries.json')
data = json.load(f)

# Create list of countries
country_list = [i['name'] for i in data]
len(country_list), country_list
```

```python
# country_list = ['Canada', 'Singapore', 'Malaysia']
c_list = []

for c in country_list:
    info = CountryInfo(c).info()
    c_dict = {'Name': info['name'], 
              'Features': 
                  {'Area': info['area'], 
                   'LatLng': info['latlng'], 
                   'Region:': info['region'], 
                   'Subregion:': info['subregion'], 
                   'Population:': info['population'], 
                   'Borders:': info['borders']}}
    c_list.append(c_dict)
    
c_list
```
