```python
# Load country json
# Includes only countries recognized by the US Department of State
f = open('countries.json')
data = json.load(f)

# Create list of countries
country_list = [i['name'] for i in data]
len(country_list), country_list
```

```python
# Format:
# {'Name': 
#  {0: 'Canada', 
#   1: 'Singapore', 
#   2: 'Malaysia'},
#  'Features': {0: {'Canada Features'}}
# }

# c_list = []

# country_list = ['Canada', 'Singapore', 'Malaysia']


name_dict = {}
features_dict = {}
for i in range(len(country_list)):
    c = country_list[i]
    info = CountryInfo(c).info()
   
    # Add to Name Dictionary
    name_dict[i] = info['name']
    
    # Add to Features Dictionary
    c_dict = {
        'Area': info['area'], 
        'LatLng': info['latlng'], 
        'Region': info['region'], 
        'Subregion': info['subregion'], 
        'Population': info['population'], 
        'Borders': info['borders']}
    features_dict[i] = c_dict
    

country_json = [{
    'Name': name_dict, 
    'Features': features_dict, 
    'nameList': country_list
}]

print(len(country_json[0]['Name']))
pprint.pp(country_json)
```

```python
names = [i['Name'] for i in c_list]
regions = [i['Features']['Region'] for i in c_list]
subregions = [i['Features']['Subregion'] for i in c_list]

print('Names:\n', names, '\n')
print('Unique Regions:\n', list(set(regions)), '\n')
print('Unique Subregions:\n', list(set(subregions)))
```
